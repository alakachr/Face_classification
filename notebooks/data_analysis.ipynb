```python
import json
import pandas as pd
```

```python
with open ("/data/ubuntu/Face_classification/data/im_name2label_train.json", "r") as f:
    data = json.load(f)

list_dict = [{"im_name": key, "label": value} for key, value in data.items()]
df = pd.DataFrame(list_dict)
df.label.value_counts().plot(kind='bar')
```

```python
df.label.value_counts()
```

```python
df_0 = df[df.label == 0]
df_1 = df[df.label == 1]
df_1_sampled = df_1.sample(n=len(df_0), random_state=42)
df_balanced = pd.concat([df_0, df_1_sampled])
```

```python
df_balanced.label.value_counts().plot(kind='bar')
```

```python
dict_balanced = {row.im_name: row.label for row in df_balanced.itertuples()}
with open ("/data/ubuntu/Face_classification/data/im_name2label_train_balanced.json", "w") as f:
    json.dump(dict_balanced, f)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path

# List of 10 image paths
root_folder =Path( "/data/ubuntu/Face_classification/data/images")
im_paths = [root_folder/im_name for im_name in  df_0.sample(10).im_name.tolist() ]
# Function to plot images in a grid
def plot_image_grid(im_paths, grid_size=(2, 5)):
    fig, axs = plt.subplots(grid_size[0], grid_size[1], figsize=(15, 6))  # Adjust the size based on your grid
    axs = axs.flatten()  # Flatten the axis array to iterate over easily
    
    for i, img_path in enumerate(im_paths):
        img = Image.open(img_path)
        axs[i].imshow(img)
        axs[i].axis('off')  # Hide the axes
    
    plt.tight_layout()
    plt.show()

# Call the function with a 2x5 grid
plot_image_grid(im_paths, grid_size=(2, 5))
```
